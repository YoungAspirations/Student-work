# Week 2 Results

```python
from numpy import loadtxt as load, unique, poly1d, polyfit
from pylab import plot, legend, xlabel, ylabel, title
import matplotlib.pyplot as plt
Temperature = [10, 250, 500, 750, 1000, 1250, 1500]
Volume_Glass = [10101, 1.0158E+04, 10230, 1.0299E+04 , 10314, 1.0329E+04, 10342]
Volume_NaCl = [11377, 1.1269E+04, 11524, 1.1969E+04, 12620,1.9625E+04, 23457]
error_GV = [1.0074E+01, 3.0489E+01, 4.4479E+01, 4.5637E+01, 4.6087E+01, 5.0501E+01, 4.6687E+01]
error_NV = [2.3093E+00, 4.3517E+00, 7.3580E+00, 2.2339E+01, 2.8695E+01,6.4726E+02, 1.1543E+03]
plt.subplot(2,1,1)
plot(Temperature, Volume_Glass, 'ro',label ='Glass')
plot(unique(Temperature), poly1d(polyfit(Temperature, Volume_Glass, 1))(unique(Temperature)))
plt.errorbar(Temperature, Volume_Glass, error_GV, linestyle = 'None', marker = '^' )
xlabel('Temperature (K)')
ylabel('Volume (Ã…Â³)')
title('Fluctuation of volume of 2 materials at different temperatures')
legend()

plt.subplot(2,1,2)
plot(Temperature, Volume_NaCl, 'bo', label ='NaCl')
plot(unique(Temperature), poly1d(polyfit(Temperature, Volume_NaCl, 1))(unique(Temperature)))
plt.errorbar(Temperature, Volume_NaCl, error_NV, linestyle = 'None', marker = '^' )
xlabel('Temperature (K)')
ylabel('Volume (Ã…Â³)')
legend()

alpha_continuousGlass = (1/Volume_Glass[0])*((Volume_Glass[3]-Volume_Glass[0])/(Temperature[3]-Temperature[0]))
alpha_continuousNaCl = (1/Volume_NaCl[0])*((Volume_NaCl[3]-Volume_NaCl[0])/(Temperature[3]-Temperature[0]))
print(alpha_continuousGlass, alpha_continuousNaCl)
#Take gradient of linear reason
```
