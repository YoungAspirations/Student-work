## Methods of root finding and their algorithms

The following code will aim to reproduce; bisection, regular falsi, and the Newton-Raphson method with results. In addition, an analysis on the efficacy of the methods will be conducted with the aid of graph plots. 

```python
## Bisection method
def bisection_step(f, bounds): #The variables are: the function f constituting a singule value and bounds, containing the range of values.  
    lower, upper = bounds      # Seperate the tuple so that the bounds can be manipulated. 
    midpoint = (f(lower) + f(upper))/2 # Calculate the midpoint
    
    if f(lower) * f(midpoint) < 0: # A conditional statement to see if a root is within the region.
        return (lower, midpoint)
    else:                      
        return (midpoint, upper) # If the first condition is not met and there is no root, then the bounds are changed to this.
    #bisection bad with 1/x type, used for continuous 
## Regula falsi
def regula_falsi_step(f, bounds):
    lower, upper = bounds
    midpoint = ((lower * f(upper)) - (upper * f(lower))) / (f(upper) - f(lower)) # The mathematical equation to calculate the regula falsi.
    if  f(lower) * f(midpoint) < 0:
        return (lower, midpoint) # The same conditional statements as stated for the bisection step
    else:
        return (midpoint, upper)
#The Regula Falsi method can only be used to find a real root in an interval [a,b] in which f[x] changes sign
## Newton's-Rapson
def newton_step(f, fp, x0): #For the Newton's-Rapson method fp stands for the first derivative and x0 the initial value.
    x1 = x0 - (f(x0) / fp(x0))
    return x1


from matplotlib.pyplot import plot, figure, xlim, ylim, grid 
from pylab import linspace
from numpy import sqrt, cos

def f(x):
    return x**2 - 4
def g(x):
    return x**8 - 2*x - 1
def h(x):
    return (x - 1)/ (x - 2)
def j(x):
    return x**3 + 47*(x**2) - 148*x + 90
def k(x):
    return x**4 - 8*(x**3) + 22*(x**2) - 24*x + 9
def l(x):
    return cos(x)
x = linspace(-6,100,500)
xlim(-10,60)
ylim(-40,50)
grid()
plot(g(x),'r-.')
plot(h(x),'k--')
plot(j(x),'m')
plot(k(x),'g')
plot(l(x),'c')
#Now, calculating some roots

#Roots using bisection_step
lower, upper = 0, 4 
steps = 0
while upper - lower > 1e-8: 
    lower, upper = bisection_step(f, (lower, upper))
    steps += 1
print(steps, lower) #good for linear continuous functions

lower, upper = 1.3,2.5 
steps = 0
while abs(upper - lower) > 1e-8: # A continuoes loop that runs until the magnitude of the difference is less than 10^-8
    lower, upper = bisection_step(h, (lower, upper))
    steps += 1
print(steps,lower) 

#Roots by regula falsi
# crashes for h(x) in range(1.3, 2.5)

lower, upper = 32, 37
s = 0
while abs(j(lower)) > 1e-8 and abs(j(upper)) > 1e-8: #
    s += 1
    lower, upper = regula_falsi_step(j, (lower, upper))
print(s, lower)

# Keeps on running with the star

```
